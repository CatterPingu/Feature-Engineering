## Count or Frequency Encoding

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',header=None,index_col=None) 
train_set.head()
```

```python
columns=[1,3,5,6,7,8,9,13]
```

```python
train_set=train_set[columns]
```

```python
train_set.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
```

```python
train_set.head()
```

```python
for feature in train_set.columns[:]:
    print(feature,":",len(train_set[feature].unique()),'labels')
```

```python
country_map=train_set['Country'].value_counts().to_dict()
```

```python
train_set['Country']=train_set['Country'].map(country_map)
```

```python
train_set.head(20)
```

### Advantages
1. Easy to use
2. No increase in feature space

### Disadvantages
1. It will not handle values having the same weight


# Taget Guided Ordinal Encoding
1. Ordering the labels according to the target
2. Replace the labels by the joint probability of being 1 or 0

```python
import pandas as pd
df=pd.read_csv('titanic_train.csv',usecols=['cabin','survived'])
df.head()
```

```python
df['cabin'].fillna('Missing',inplace=True)
```

```python
df.head(20)
```

```python
df['cabin']=df['cabin'].astype(str).str[0]

#Capturing only the coach and removing the seat numbers
```

```python
df['cabin'].unique()
```

```python
df.groupby(['cabin'])['survived'].mean()

#In each cabin the percentage of people surviving
```

```python
df.groupby(['cabin'])['survived'].mean().sort_values()
```

```python
ordinal_labels=df.groupby(['cabin'])['survived'].mean().sort_values().index

print(ordinal_labels)
```

```python
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,0)}

#Iterating through the labels and assigning from zero(dictionary comprehension)

ordinal_labels2
```

```python
df['cabin_ordinal_label']=df['cabin'].map(ordinal_labels2)
```

```python
df.head(20)
```

Here, the target was found i.e. rank was found through the probability of survival and not just on the basis of frequency of the value in a category

# Mean Encoding

Similar to target encoding but here instead of assigning ranks according to the target, the encoding is done on the with the mean obtained from the target

```python
df.groupby(['cabin'])['survived'].mean()
```

```python
df.groupby(['cabin'])['survived'].mean().to_dict()
```

```python
mean_ordinal=df.groupby(['cabin'])['survived'].mean().to_dict()
```

```python
mean_ordinal
```

```python
df['mean_ordinal_label']=df['cabin'].map(mean_ordinal)
```

```python
df.head(20)
```

### Advantages
1. Captures information within the label i.e. rendering more predictive features
2. Creates a monotonic relationship between the label and the target

### Disadvantages
1. Prone to overfitting
